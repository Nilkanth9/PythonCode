Use Case 1: Clean Employee Data
We have a dataset with missing or invalid values like:
None
NaN
empty strings ""
Our goal is to remove those invalid entries and return a clean version of the data.

```python
import pandas as pd

def clean_data(data):
    cleaned_dict = {}

    for key, values in data.items():
        cleaned_list = []
        for value in values:
            if isinstance(value, list):
                inner_clean = [v for v in value if not pd.isna(v)]
                cleaned_list.append(inner_clean)

            elif not pd.isna(value):
                cleaned_list.append(value)
        cleaned_dict[key] = cleaned_list
    return cleaned_dict
```

```python
updated_dict = clean_data(data)
print(updated_dict)
```

```python
def Clean_data1(data):
  for i,j in data.items():
    for val in range(len(j)):
      if isinstance(j[val],list):
        j[val]=[n for n in j[val] if n is not None]
      elif j[val] is None:
        j[val]="Nil"
  return data
```

```python
def clean_data1(data):
    return {
        new: [
            list(filter(lambda x: x is not None,val)) if isinstance(val,list) else(val if val is not None else "nothing")
            for val in new_dict
        ]
        for new,new_dict in data .items()
    }
```

```python
def clean2(data):
  clean={}
  for key,value in data .items():
     clean_list=[]
     for val in value:
      if isinstance(val,list):
        inner_list_list=[v for v in val if not pd.isna(v)]
        clean_list.append(inner_list_list)
      elif not pd.isna(val):
        clean_list.append(val)
     clean[key]=clean_list
  return clean
```

```python
def clean_data3(data):
    return {
        key: [
            list(filter(lambda x: x is not None, val)) if isinstance(val, list)
            else (val if val is not None else "nothing") if not isinstance(val, dict)
            else {k:v for k,v in val.items() if v is not None}  # clean dictionary
            for val in values
        ]
        for key, values in data.items()
    }
```

Use Case Calculate Avg salary Per dept

```python
def calculate_avg(data):
  dict={}
  for d,s in zip(data["Department"],data["Salary"]):
    if not d or not s or str(d) == 'nan' or str(s) == 'nan':
       continue
    dict.setdefault(d,[]).append(s)
  return{d:sum(v)/len(v)for d,v in dict.items()}

```

```python
calculate=calculate_avg(data)
print(calculate)
```

Sort Emp By salary

```python
def sort_by_salary(data):
  emp=list(zip(data["Name"],data["Salary"]))
  emp.sort(key=lambda x:x[1] if x[1] is not None else float('inf'))
  return emp


```

```python
result = sort_by_salary(data)

for name, sal in result:
    print(name, sal)
```

Same sorting using simple function

```python
def salary_key(x):
    return x[1] if x[1] is not None else float('inf')

def sort_by_salary(data):
    emp = list(zip(data["Name"], data["Salary"]))
    emp.sort(key=salary_key)
    return emp
```

```python
new_sort=sort_by_salary(data)
print(new_sort)
```

Count Emp per dept

```python
def emp_per_dept(data):
  emp_dept={}
  for i in data["Department"]:
    if isinstance(i,str):
      emp_dept[i]=emp_dept.get(i,0)+1
  return emp_dept
```

```python
cot=emp_per_dept(data)
print(cot)
```

find employees whose salary is above the companyâ€™s average salary.

```python
def emp_above_company_avg(data):
    valid_salary=[x for x in data["Salary"] if x is not None]
    avg_salary=sum(valid_salary)/len(valid_salary)
    emp_above_avg=[]
    for name,sal in zip(data["Name"],data["Salary"]):
      if sal is not None and sal>avg_salary:
        emp_above_avg.append(name)
    return{"Company_avg_sal":avg_salary,"Above_avg_emp":emp_above_avg}
```

```python
above=emp_above_company_avg(data)
print(above)
```

Genrate Department Summery report

```python
df = pd.DataFrame(data)
```

```python
def dept_summary(df):
    return df.groupby("Department")[["Salary", "RemoteWork"]].agg(
        Employee_Count=("Salary", "count"),
        Avg_Salary=("Salary", "mean"),
        Remote_Workers=("RemoteWork", "sum"),
        maximum_sal=("Salary","max"),
        minumum_sal=("Salary","min")

    )
```

```python
summery=dept_summary(df)
summery
```
